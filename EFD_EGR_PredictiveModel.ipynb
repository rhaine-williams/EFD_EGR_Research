```python
#Dropping all NaNs from Data (significantly reduced to 136)
df1=df1.dropna(axis=0)
df1
```

```python
df1=df1.drop(columns=["Student ID"])
df1
```

```python
#df1 = df1.drop(subset=['120-1'], how='F')
df1.drop(df1[df1['120-1'] == 'F'].index, inplace = True)
df1.drop(df1[df1['120-1'] == 'W'].index, inplace = True)
df1.drop(df1[df1['100-1'] == 'U'].index, inplace = True)
df1
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
feature_array = ohe.fit_transform(df1[['100-1','Gender','Ethnicity','Military Status','International']]).toarray()
feature_array.ndim
```

```python
labels = np.array(ohe.categories_,dtype=object)
feature_labels = np.concatenate((labels[0],labels[1],labels[2],labels[3],labels[4]))
feature_labels
```

```python
features = pd.DataFrame(feature_array, columns = feature_labels)
features
```

```python
df1=df1.reset_index(drop=True)
```

```python
df2 = df1.join(features)
df2
```

```python
df_final=df2.drop(['100-1','Gender','Ethnicity','Military Status','International'], axis=1)
```

```python
#Target Variable (y)
y = df_final["120-1"]

#Features (X)
df_final=df_final.drop(['120-1'], axis=1)
X = df_final
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import LinearSVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
seed = 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.4, stratify = y, random_state = seed)
```

```python
#scaler = MinMaxScaler()
#X_train = scaler.fit_transform(X_train)
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
ypred_train = clf.predict(X_train)

# Evaluating Accuracy:
print('The accuracy on the training set is: ',accuracy_score(y_train, ypred_train))
```

```python
ypred_test = clf.predict(X_test)

# Evaluating Accuracy:
print('The accuracy on the testing set is: ',accuracy_score(y_test, ypred_test))
```

```python
clf.classes_
```

**RANDOM FOREST**

```python
random_forest = RandomForestClassifier(n_estimators=100, random_state=0)
random_forest.fit(X_train, y_train)
```

```python
RF_pred_train = random_forest.predict(X_train)
```

```python
print('The accuracy on the training set is: ',accuracy_score(y_train, RF_pred_train))
```

```python
RF_pred_test = random_forest.predict(X_test)
RF_pred_test
```

**ACCURACY**

```python
print('The accuracy on the testing set is: ',accuracy_score(y_test, RF_pred_test))
```

**CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, RF_pred_test,labels=random_forest.classes_)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=random_forest.classes_)
disp.plot()

plt.show()
```

**PRECISION**

```python
from sklearn.metrics import precision_score
precision_score(y_test, RF_pred_test, average='weighted')
```

**F1 SCORE**

```python
from sklearn.metrics import f1_score
f1_score(y_test, RF_pred_test, average='weighted')
```

**RECALL**

```python
from sklearn.metrics import recall_score
recall_score(y_test, RF_pred_test, average='weighted')
```
